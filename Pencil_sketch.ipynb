```
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```
import numpy as np
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```
img ="https://1.bp.blogspot.com/-l783Guy8tqM/Ui9Zd4wnbbI/AAAAAAAAfB8/zoJyaqhZRpI/s1600/151904392_1358487195.jpg"
```

```
import imageio
s = imageio.imread(img)
g=grayscale(s)
i = 255-g
import scipy.ndimage
b = scipy.ndimage.filters.gaussian_filter(i,sigma=100)
r= dodge(b,g)
```

```
import matplotlib.pyplot as plt
plt.title('Sketch', fontsize=16)
plt.imshow(r, cmap="gray")

plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```
plt.imsave('sketch.png', r, cmap='gray', vmin=0, vmax=255)
```
